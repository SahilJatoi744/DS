```python
# Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
import scipy
```

```python
# dataset download

kashti = sns.load_dataset("titanic")
kashti.head(10)
```

```python
kashti.isna().sum()
```

```python
kashti.dropna(subset=['age'],axis=0, inplace=True)
kashti.isna().sum()
```

```python
df = kashti[['sex', 'age', 'fare']]
df.head()
```

```python
# ttest to compare the age of male vs female

#1- import libraries
from scipy.stats import ttest_ind
#2-sub sets of male vs female
df_male = df[df['sex']=="male"]
df_female = df[df['sex'] == "female"]
#3- ttest (unpaired, two sample / pr independent ttest)
ttest_ind(df_male['age'],df_female['age'])
stat, p_value = ttest_ind(df_male['age'], df_female['age'])
print("stat", stat, "p_value", p_value)
#4- conditional loop , different or not?

if p_value > 0.05:
    print("There is no significant difference")
else:
    print("There is a significant difference")
```

```python
# One sample ttest

#1- import libraries
from scipy.stats import ttest_1samp
#2-sub sets of male vs female
df_male = df[df['sex']=="male"]
#3- ttest (unpaired, two sample / pr independent ttest)
ttest_1samp(df_male['age'],36)
stat, p_value = ttest_1samp(df_male['age'], 36)
print("stat", stat, "p_value", p_value)
#4- conditional loop , different or not?

if p_value > 0.05:
    print("There is no significant difference")
else:
    print("There is a significant difference")
```

### Work on it

```python
# paired ttest

#1- import libraries
from scipy.stats import ttest_ind
#2-sub sets of male vs female
df_male = df[df['sex']=="male"]
df_female = df[df['sex'] == "female"]
#3- ttest (unpaired, two sample / pr independent ttest)
ttest_ind(df_male['age'],df_female['age'])
stat, p_value = ttest_ind(df_male['age'], df_female['age'])
print("stat", stat, "p_value", p_value)
#4- conditional loop , different or not?

if p_value > 0.05:
    print("There is no significant difference")
else:
    print("There is a significant difference")
```
