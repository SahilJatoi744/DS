```python
import plotly.express as px
import pandas as pd
df = pd.read_csv("train.csv")
```

```python
df.columns
```

```python
df.info()
```

```python
fig = px.scatter(x=df['battery_power'], y=df['price_range'],
	         size=df["ram"], color=df['n_cores'],
                 hover_name=df["n_cores"], log_x=True, size_max=60)
fig.show()
```

```python
corr = df.corr(method="pearson")
corr
```

#### PLOTS

```python
fig = px.violin(x=df["three_g"], y=df["four_g"], box=True, # draw box plot inside the violin
               )
fig.show()
```

```python
fig = px.violin(x=df["talk_time"], y=df["battery_power"],color=df["n_cores"], box=True # draw box plot inside the violin
               )
fig.show()
```

```python
fig = px.violin(x=df["touch_screen"], y=df["n_cores"],color=df["blue"], box=True # draw box plot inside the violin
               )
fig.show()
```

```python
fig = px.violin(x=df["wifi"], y=df["px_height"], box=True # draw box plot inside the violin
               )
fig.show()
```

```python
fig = px.violin(x=df["price_range"], y=df["ram"],color=df["n_cores"], box=True # draw box plot inside the violin
               )
fig.update_traces(box_visible=True, meanline_visible=True)
fig.update_layout(violinmode='group')
fig.show()
```
