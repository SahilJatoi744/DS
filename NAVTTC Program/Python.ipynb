```python
import random
l = int(input("Enter number 6 or 10"))
name = input("What is your name? ")
print("Good Luck ! ", name)
lf6 = ['access', 'apple', 'accurate']
lf10 = ['helloworld']
l1 = ['_', '_', '_', '_', '_', '_']
print("Guess the characters")
guesses = ''
turns = 10
while turns > 0:
    failed = 0
    if l == 10:
        word = random.choice(lf10)
    else:
        word = random.choice(lf6)
        for char in word:
            if char in guesses:
                print(char, l1)
            else:
                print(end=" ")
                failed += 1
        if failed == 0:
            print("You Win")
            print("The word is: ", word)
            break
        print()
        guess = input("guess a character:")
        guesses += guess
        if guess not in word:
            turns -= 1
            print("Wrong")
            print("You have", + turns, 'more guesses')
            if turns == 0:
                print("You Loose")
```
